```python
import pandas as pd
```

```python
data = pd.read_csv('raw.csv')
```

```python
data.head()
```

```python
filtered = data[data['channel_title'].str.contains('VEVO')]

ft = filtered['title'].str.contains('ft.')
coder = {True:False, False:True}
ft = ft.map(coder)
filtered = filtered[ft]
filtered
```

```python
artist=[]
song=[]

for title in filtered['title']:
    artist.append(title[0:title.find(' - ')])
    if (title.find(' (Official Video)')>-1):
        song.append(title[title.find(' - ')+3:title.find(' (Official Video)')])
    else:
        song.append(title[title.find(' - ')+3:])
```

```python
finished = pd.DataFrame({
    'video_id': filtered['video_id'],
    'artist': artist,
    'song': song
})
```

```python
finished
```

```python
finished.to_csv('links.csv', encoding='utf-8-sig')
```
